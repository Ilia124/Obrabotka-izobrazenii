```python
import numpy as np
import cv2
import os
import argparse
import matplotlib.pyplot as plt
```

```python
# медианный фильтр
ap = argparse.ArgumentParser()
ap.add_argument('-f', '-i' '--image', help = '/content/1.jpg')
args = vars(ap.parse_args())
# загружаем изображение
image = cv2.imread('/content/1.jpg')
# применяем медианный фильтр 3 × 3 к изображению
processed_image = cv2.medianBlur(image, 3)
# Выводим изображение
# Сохраняем изображение в папку с проектом
cv2.imwrite('result1.jpg', processed_image)
plt.imshow(cv2.cvtColor(processed_image, cv2.COLOR_BGR2RGB))
```
